```python
# import libraries
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
from sqlalchemy import create_engine
```

```python
def get_URI(query:str, page_num:str, begin_date:str, end_date:str, API_KEY:str) -> str:
    """# returns the URL to the articles for the current query by page number and date """

    # add request to uri
    URI=f'https://api.nytimes.com/svc/search/v2/articlesearch.json?q={query}'

    # add page number and date
    URI = URI + f'&page={page_num}&begin_date={begin_date}&end_date={end_date}'

    # add API key
    URI = URI + f'&api-key={API_KEY}'

    return URI
```

# Extract

```python
import time
import datetime

# create a dataframe to store all records
df = pd.DataFrame()

# get the current date
begin_date = (datetime.datetime.now() - datetime.timedelta(5)).strftime('%Y%m%d')
end_date = datetime.datetime.now().strftime('%Y%m%d')

# collect data from all available pages
page_num = 1

# load settings
with open('settings.json', 'r') as f:
     settings = json.load(f)

while True:

# get the URI with entries related to the Winter Olympic Games on the current date
     URI = get_URI(query='COVID', page_num=str(page_num), begin_date=begin_date, end_date=end_date, API_KEY=settings['API_KEY'])

# make a request by URI
     response = requests.get(URI)

# convert result to JSON format
     data = response.json()

# transform the data into a dataframe
     df_request = json_normalize(data['response'], record_path=['docs'])

# break the loop if there are no new entries
     if df_request.empty:
         break

     # add records to the end of the date frame
     df = pd.concat([df, df_request])

     # pause for request by number of requests
     time.sleep(6)

     # go to the next page
     page_num += 1
```

# Transform

```python
# look for duplicates and remove them
if len(df['_id'].unique()) < len(df):
     print('There are duplicates in the data')
     df = df.drop_duplicates('_id', keep='first')

# find and remove entries without titles
if df['headline.main'].isnull().any():
     print('There are missing values in this dataset')
     df = df[df['headlinee.main'].isnull()==False]

# filter op-ed articles
df = df[df['type_of_material']!='op-ed']

# leaving only the headline, publication_date, author name and url fields
df = df[['headline.main', 'pub_date', 'byline.original', 'web_url']]

# rename columns columns
df.columns = ['headline', 'date', 'author', 'url']
```

# Load

```python
# create an engine object for the database
username = settings['DB']['username']
password = settings['DB']['password']
database = settings['DB']['database']
schema = settings['DB']['schema']
database_loc = f"postgresql://{username}:{password}@localhost:5432/{database}"
engine = create_engine(database_loc)

# Adding data to the database
df.to_sql(name='news_articles',
          con=engine,
          index=False,
          if_exists='append',
          schema=schema)
```
